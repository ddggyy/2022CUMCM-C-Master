```python
import pandas as pd
data3 = pd.read_excel('整合数据.xlsx',sheet_name='表单3')
data3=pd.concat([data3,pd.get_dummies(data3['表面风化'])],axis=1)
#data3.head(5)
```

```python
data3.fillna(0.04,inplace=True)
data3
```

```python
data_merage = pd.read_excel('整合数据.xlsx',sheet_name='表单2')
data_merage=data_merage[['类型','表面风化','二氧化硅(SiO2)', '氧化钠(Na2O)',
       '氧化钾(K2O)', '氧化钙(CaO)', '氧化镁(MgO)', '氧化铝(Al2O3)', '氧化铁(Fe2O3)',
       '氧化铜(CuO)', '氧化铅(PbO)', '氧化钡(BaO)', '五氧化二磷(P2O5)', '氧化锶(SrO)',
       '氧化锡(SnO2)', '二氧化硫(SO2)']]

data_merage=pd.concat([data_merage,pd.get_dummies(data_merage['表面风化'])],axis=1)
```

```python
data_merage.columns
```

```python
import xgboost as xgb
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import classification_report,f1_score
from sklearn.model_selection import train_test_split
```

```python
X=data_merage[[ '二氧化硅(SiO2)', '氧化钠(Na2O)', '氧化钾(K2O)', '氧化钙(CaO)',
       '氧化镁(MgO)', '氧化铝(Al2O3)', '氧化铁(Fe2O3)', '氧化铜(CuO)', '氧化铅(PbO)',
       '氧化钡(BaO)', '五氧化二磷(P2O5)', '氧化锶(SrO)', '氧化锡(SnO2)', '二氧化硫(SO2)', '无风化',
       '风化']]
y=data_merage['类型']
X.fillna(0.04,inplace=True)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf1 = LogisticRegression()
clf1.fit(x_train, y_train)
y_pred = clf1.predict(x_test)
accuracy1 = balanced_accuracy_score(y_test,y_pred)
print(round(accuracy1,3))
```

```python
clf2 = RandomForestClassifier()
clf2.fit(x_train, y_train)
y_pred = clf2.predict(x_test)
accuracy2 = balanced_accuracy_score(y_test,y_pred)
print(round(accuracy2,3))
```

```python
clf3 = lgb.LGBMClassifier()
clf3.fit(x_train, y_train)
y_pred = clf3.predict(x_test)
accuracy3 = balanced_accuracy_score(y_test,y_pred)
print(round(accuracy3,3))
```
