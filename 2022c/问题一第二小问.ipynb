```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_style('whitegrid',{'font.sans-serif':['simhei','Arial']})
```

```python
df = pd.read_excel('整合数据.xlsx',sheet_name='高钾风化')
df.replace(0, 0.04, inplace=True
           
```

```python
df.columns
```

```python
columns_to_plot = [  '二氧化硅(SiO2)','氧化钠(Na2O)','氧化钾(K2O)', '氧化钙(CaO)', '氧化镁(MgO)', '氧化铝(Al2O3)', '氧化铁(Fe2O3)',
       '氧化铜(CuO)', '氧化铅(PbO)', '氧化钡(BaO)', '五氧化二磷(P2O5)', '氧化锶(SrO)','氧化锡(SnO2)', '二氧化硫(SO2)']
```

```python
data_to_plot = df[columns_to_plot]
```

```python
# 定义归一化函数
def normalize_data(col):
    return (col - col.min()) / (col.max() - col.min())
```

```python

normalized_data = data_to_plot.apply(normalize_data)
```

```python
sns.boxplot(data=normalized_data, orient="h")
plt.tight_layout()
plt.savefig('Features\\高钾风化箱线图.pdf')
```
